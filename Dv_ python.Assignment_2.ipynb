## 2: Plot 12 most dangerous cities

```python
city_freq = kill.city.value_counts()[0:12]
city_freq
```

```python
# with seaborn barplot

fig, ax =plt.subplots(figsize=(8,5))

sns.barplot(x=kill.city.value_counts()[0:12].index, y=kill.city.value_counts()[0:12]).set(title='Most dangerous cities')

plt.xticks(rotation=35);
```

```python
# with seaborn countplot
fig, ax =plt.subplots(figsize=(8,5))
ax = sns.countplot(x = 'city',
              data = kill,
              order = kill['city'].value_counts()[0:12].index).set(title='Most dangerous cities')
plt.xticks(rotation=35);
```

## 3: Plot 20 most dangerous states

```python
state_freq = kill.state.value_counts()[0:20]
state_freq
```

```python
fig, ax =plt.subplots(figsize=(8,5))

sns.barplot(x=kill.state.value_counts()[0:20].index,
            y=kill.state.value_counts()[0:20]).set(title='Most dangerous states')

plt.xticks(rotation=35);
```

```python
# with seaborn countplot

fig, ax =plt.subplots(figsize=(8,5))
ax = sns.countplot(x = 'state',
              data = kill,
              order = kill["state"].value_counts()[0:12].index).set(title='Most dangerous states')

plt.xticks(rotation=35);              
```
